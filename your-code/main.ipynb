## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
# your code here
for word in words:
    print(word.upper())
```

**Create a new list containing only words with 5 or more letters**

```python
# your code here
long_words = [word for word in words if len(word)>=5]
print(long_words)
```

**Print the first word starting with "t"**

```python
# your code here
for word in words:
    if word.startswith('t'):
        print(word)
        break
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
# your code here
squares = [i**2 for i in range(1,11)]
print(squares)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
# your code here
squares = [i**2 for i in range(1,11) if i%2!=0]
print(squares)
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
# your code here
squares_of_multiples_of_8 = [i**2 for i in range(8, 1000, 8)]
print(squares_of_multiples_of_8)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
# your code here
print(len(people))
```

**How many people have kids**?

```python
# your code here
print(len([person for person in people if person['n_kids']>0]))
```

**How many kids do they have in total?**

```python
# your code here
print(sum([person['n_kids'] for person in people]))
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
# your code here
for person in people:
    if person['name'].endswith('a'):
        person['n_kids'] += 1

print(people)
```
